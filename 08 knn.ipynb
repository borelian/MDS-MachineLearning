# Función train_test_split

Primero separaremos los datos en train y test, más detalles en el próximo tópico. Para ello usaremos la función `train_test_split` de sklearn.model_selection.

```
train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True)
```

* *arrays: bases de datos con los datos a separar.
* test_size: float entre 0 y 1 y representa el porcentaje de los datos que serán usados como test data. 1-train_size.
* train_size: float entre 0 y 1 y representa el porcentaje de los datos que serán usados como train data. Si se deja vacío es 1-test_size.
* random_state: semilla para poder repetir el experimento.
* shuffle: Si los datos deben ser mezclados antes de la separación.

Ejemplo:<br>
```from sklearn.model_selection import train_test_split<br>
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.33, random_state=42)
```

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target, test_size=0.33,random_state=42)

#Entrenando el modelo con los datos de iris
trainData=pd.DataFrame(X_train,columns=iris.feature_names)
trainData["Labels"]=y_train
testData=pd.DataFrame(X_test,columns=iris.feature_names)
testData["Labels"]=y_test
````

# Clase KNeighborsClassifier

Existen múltiples parámetros para el modelo KNN:<br>
`KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='minkowski', p=2, metric_params=None)`<br>
Parámetros
* n_neighbors: número de vecino a utilizar
* weights: pesos a utilizar en la clasificación (‘uniform’, ‘distance’: mayor importancia a los puntos cercanos)
* metric: Métrica de distancia a utilizar entre los puntos $\mathbf x_1$ y $\mathbf x_2$, cada uno con d dimensiones<br>
    "euclidean" $\Rightarrow\sqrt{\sum_{i=1}^d(x_{1,d} - x_{2,d})^2}$<br>
    "manhattan" $\Rightarrow\sum_{i=1}^d|x_{1,d} - x_{2,d}|$<br>
    "chebyshev” $\Rightarrow\max_{i\in\{1\ldots d\}}|x_{1,d} - x_{2,d}|$<br>
    "MinkowskiDistance", con parametro p $\Rightarrow\left(\sum_{i=1}^d(x_{1,d} - x_{2,d})^p\right)^{1/p}$<br>
    "mahalanobis", con parametro V $\sqrt{(\mathbf x_1-\mathbf x_2)^TV^{-1}(\mathbf x_1-\mathbf x_2)}$
* p: parámetro de la métrica de Minkowsky (2 es equivalente a la distancia Euclidiana)
* metrics_params: parámetros específicos de la métrica
* algorithm: algoritmo para buscar los vecinos más cercanos, en caso de Mahalanobis debe ser "brute"

<br>Atributos
* classes_: Etiquetas de los puntos conocidos
* effective_metric_: Métrica de distancia
* n_samples_fit_: número de puntos de "entrenamiento" 

<br>métodos/funciones
* fit(X, y): "Entrenamiento" del modelo, se tiene que dar los datos y la clase Y
* kneighbors([X, n_neighbors, return_distance]): Muestra los K-vecinos de un punto.
* predict(X): predice las etiquetas para los puntos dados
* predict_proba(X): predice las probabilidades de las clases para los puntos dados

````python
#Creando un objeto KNN con las condiciones iniciales
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
````

````python

#Entrenando el modelo con los datos de iris
KNN=KNN.fit(X_train,y_train)
````

````python
KNN.score(X_train,y_train)
````

````python
KNN.score(X_test,y_test)
````

````python
#Aplicando el modelo a otros datos
resultado=KNN.predict(X_test)
````

# Ejemplo con datos de distintas escalas

````python
#Generando dos lunas, pero re-escalando los indices
from sklearn.datasets import make_moons
tempData=make_moons(n_samples=1000,shuffle=True, noise=0.3, random_state=None)
tempData[1]
tempData[0][:,0]=tempData[0][:,0]*0.01 # dividimos por 100
tempData[0][:,1]=tempData[0][:,1]*2    # multiplicamos por 2

tempDF=pd.DataFrame(tempData[0],columns=["X1","X2"])
tempDF["label"]=tempData[1]

(ggplot(tempDF)
 +aes(x="X1",y="X2",color="factor(label)")
 +geom_point(show_legend=False, alpha=0.8)
 #+xlim(-3,3) + ylim(-3,3)
)
````

````python
#Entrenando el modelo y aplicando a una grilla de puntos para su visualización
KNN = KNeighborsClassifier(n_neighbors=5)
KNN=KNN.fit(tempData[0],tempData[1])

#Creando una grilla de evaluación
x = np.arange(-2.0, 3.5, 0.05)*0.01
y = np.arange(-1.5, 2.5, 0.05)*2
temp = np.meshgrid(x, y)
dataTest=pd.DataFrame(temp[0].flatten(),columns=["X1"])
dataTest.insert(1,"X2",temp[1].flatten())

#Prediciendo y graficando el output
estimatedProbs=KNN.predict_proba(dataTest.iloc[:,0:2])

#Creando el dataframe a graficar
output=(((1-estimatedProbs[:,0])*255).astype("int"))
dataTest.insert(2, "ClassOutput", "0")
for i in range(dataTest.shape[0]):
    dataTest.ClassOutput[i]='#%02x%02x%02x' % (255-output[i],0,output[i])

#Colores para los puntos originales
colorData=pd.DataFrame(list(map(str,tempData[1])),columns=["colorClass"])
colorData.colorClass[colorData.colorClass=="0"]="Red"
colorData.colorClass[colorData.colorClass=="1"]="Blue"
    
# Generating the plot with ggplot/plotnine
pp=(ggplot()+theme_void()
  +geom_tile(aes(x="X1",y="X2"),data=dataTest,alpha=0.6,show_legend=False,fill=dataTest.ClassOutput)
  +geom_point(aes(x="X1",y="X2"),data=tempDF,show_legend=False,color="black",fill=colorData)
)
print(pp)

````
